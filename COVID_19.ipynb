```python
import os
print(os.getcwd())
```

```python
import folium
from folium.plugins import MarkerCluster
```

```python
import pandas as pd
import googlemaps

hos_gps = pd.read_csv('../data/covid19_current_20210602.csv', encoding = 'euc-kr',)
gmaps_key = "> _<"
gmaps = googlemaps.Client(key=gmaps_key)

lat = []
lng = []
for name in hos_gps['주소']:
    try:
        tmpMap = gmaps.geocode(name, language='ko')
        tmpLoc = tmpMap[0].get('geometry')
        lat.append(tmpLoc['location']['lat'])
        lng.append(tmpLoc['location']['lng'])
    except:
        print("{}Address Error".format(name))
        lat.append(None)
        lng.append(None)
        
hos_gps['lat'] = lat
hos_gps['lng'] = lng
```

```python
hos_gps.dropna(subset=['lat'], inplace = True)
hos_gps
```

```python
hos_gps.isnull()
```

```python
mapping=folium.Map(location=[hos_gps['lat'].mean(),hos_gps['lng'].mean()],
                  zoom_start=11)
marker_cluster = MarkerCluster().add_to(mapping)
```

```python
map = folium.Map(location=[37.5502, 126.982],
                zoom_start=10.5,
                tiles='Stamen Toner')

from folium.plugins import MarkerCluster
marker_cluster = MarkerCluster().add_to(map)



for n in hos_gps.index:
    if pd.notnull(hos_gps['lat'][n]):
        folium.Marker([hos_gps['lat'][n], hos_gps['lng'][n]],
                      tooltip = hos_gps['시설명'][n],
                      radius=5, color='#CD3181',
                           fill_color='#CD3181',
                           fill=True).add_to(marker_cluster)
            
map            

map.save('../extract_img/hos_gps.html')
```

## 기사 댓글 크롤링

```python
List=[]
# 라이브러리를 로드합니다.
from bs4 import BeautifulSoup
import requests
import re
import sys
import pprint

# 네이버 뉴스 url을 입력합니다.
url="https://news.naver.com/main/read.nhn?mode=LSD&mid=sec&sid1=102&oid=214&aid=0001121847"

oid=url.split("oid=")[1].split("&")[0]
aid=url.split("aid=")[1]
page=1    
header = {
    "User-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.181 Safari/537.36",
    "referer":url,
    
} 
while True :
    c_url="https://apis.naver.com/commentBox/cbox/web_neo_list_jsonp.json?ticket=news&templateId=default_society&pool=cbox5&_callback=jQuery1707138182064460843_1523512042464&lang=ko&country=&objectId=news"+oid+"%2C"+aid+"&categoryId=&pageSize=20&indexSize=10&groupId=&listType=OBJECT&pageType=more&page="+str(page)+"&refresh=false&sort=FAVORITE" 
# 파싱하는 단계입니다.
    r=requests.get(c_url,headers=header)
    cont=BeautifulSoup(r.content,"html.parser")    
    total_comm=str(cont).split('comment":')[1].split(",")[0]
   
    match=re.findall('"contents":([^\*]*),"userIdNo"', str(cont))
# 댓글을 리스트에 중첩합니다.
    List.append(match)
# 한번에 댓글이 20개씩 보이기 때문에 한 페이지씩 몽땅 댓글을 긁어 옵니다.
    if int(total_comm) <= ((page) * 20):
        break
    else : 
        page+=1


# 여러 리스트들을 하나로 묶어 주는 함수입니다.
def flatten(l):
    flatList = []
    for elem in l:
        # if an element of a list is a list
        # iterate over this list and add elements to flatList 
        if type(elem) == list:
            for e in elem:
                flatList.append(e)
        else:
            flatList.append(elem)
    return flatList


# 리스트 결과입니다.
flatten(List)
```

```python
def drawKorea(targetData, blockedMap, cmapname):
    gamma = 0.75

    whitelabelmin = (max(blockedMap[targetData]) - 
                                     min(blockedMap[targetData]))*0.25 + \
                                                                min(blockedMap[targetData])

    datalabel = targetData

    vmin = min(blockedMap[targetData])
    vmax = max(blockedMap[targetData])

    mapdata = blockedMap.pivot_table(index='y', columns='x', values=targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    plt.figure(figsize=(9, 11))
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=cmapname, 
               edgecolor='#aaaaaa', linewidth=0.5)

    # 지역 이름 표시
    for idx, row in blockedMap.iterrows():
        # 광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시한다. 
        #(중구, 서구)
        if len(row['ID'].split())==2:
            dispname = '{}\n{}'.format(row['ID'].split()[0], row['ID'].split()[1])
        elif row['ID'][:2]=='고성':
            dispname = '고성'
        else:
            dispname = row['ID']

        # 서대문구, 서귀포시 같이 이름이 3자 이상인 경우에 작은 글자로 표시한다.
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 10.0, 1.1
        else:
            fontsize, linespacing = 11, 1.

        annocolor = 'white' if row[targetData] > whitelabelmin else 'black'
        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold',
                     fontsize=fontsize, ha='center', va='center', color=annocolor,
                     linespacing=linespacing)

    # 시도 경계 그린다.
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c='black', lw=2)

    plt.gca().invert_yaxis()
```

```python
from bs4 import BeautifulSoup 
from urllib.request import urlopen
import urllib
import time

tmp1 = 'https://search.naver.com/search.naver?sm=tab_hty.top&where=news' 
html = tmp1 + '&ie=utf8&query={key_word}&start={num}'

   
response = urlopen(html.format(num=1, key_word=urllib.parse.quote('아스트라제네카')))

soup = BeautifulSoup(response, "html.parser")

tmp = soup.find_all('li')
```

```python
tmp_list = []
for line in tmp:
    tmp_list.append(line.text)

tmp_list
```

```python
from tqdm import tqdm_notebook

present_candi_text = []

for n in tqdm_notebook(range(1, 1000, 10)):
    response = urlopen(html.format(num=n, key_word=urllib.parse.quote('아스트라 부작용')))

    soup = BeautifulSoup(response, "html.parser")

    tmp = soup.find_all('li')

    for line in tmp:
        present_candi_text.append(line.text)
        
    time.sleep(0.5)
```

## 크롤링해온 데이터 정제 (중복단어 및 관련없는 단어 제거)

```python
present_candi_text
```

```python
import nltk
from konlpy.tag import Twitter
t= Twitter()
```

```python
len(present_candi_text)
```

```python
present_text = ''

for each_line in present_candi_text[:len(present_candi_text)]:
    present_text = present_text + each_line + '\n'
```

```python
tokens_ko = t.nouns(present_text) #명사  ***중요!!!***
tokens_ko
```

```python
ko = nltk.Text(tokens_ko, name = '아스트라 반응')
print(len(ko.tokens))
print(len(set(ko.tokens)))  #중복 제거
```

```python
tokens_ko = [each_word for each_word in tokens_ko
                        if each_word not in stop_words]
```

```python
ko = nltk.Text(tokens_ko, name = '아스트라 반응')

ko.vocab().most_common(100)
```

## 정제한 단어를 통해 워드클라우드 작업 (기본, 이미지 활용)

```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

```python
data = ko.vocab().most_common(300)

# for win : font_path='c:/Windows/Fonts/malgun.ttf'
wordcloud = WordCloud(font_path='/Library/Fonts/malgun.ttf',
                      relative_scaling = 0.2,
                      #stopwords=STOPWORDS,
                      background_color='white',
                      ).generate_from_frequencies(dict(data))
plt.figure(figsize=(16,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
data = ko.vocab().most_common(300)

# for win : font_path='c:/Windows/Fonts/malgun.ttf'
wordcloud = WordCloud(font_path='/Library/Fonts/malgun.ttf',
                      relative_scaling = 0.2,
                      #stopwords=STOPWORDS,
                      background_color='white', mask = news_mask
                      ).generate_from_frequencies(dict(data))
plt.figure(figsize=(16,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
